## Matrix multiplication

### Pure Python implementation

````python
def mul_py(x,y):

    m1,n1 = x.shape
    m2,n2 = y.shape

    z = np.zeros((m1,n2))

    for i in range(m1):
        for j in range(n2):
            val = 0
            for k in range(m2):
                val += x[i][k] * y[k][j]
            z[i][j] = val

    return z
````

### Cythonized version

````python
%%cython --annotate
import numpy as np
cimport numpy as np
def mul_cy(np.ndarray x, np.ndarray y):

    cdef int m1 = x.shape[0]
    cdef int n1 = x.shape[1]
    cdef int m2 = y.shape[0]
    cdef int n2 = y.shape[1]
    cdef np.ndarray z = np.zeros([m1, n2], dtype=float)

    cdef int i
    cdef int j
    cdef int k

    cdef double value

    for i in range(m1):
        for j in range(n2):
            value = 0
            for k in range(m2):
                value += x[i][k] * y[k][j]

            z[i][j] = value

    return z
````

````python
N = 100
arr1 = np.random.random((N,N))
arr2 = np.random.random((N,N))
````

````python
%timeit -n 1  mul_py(arr1,arr2)
%timeit -n 1  mul_cy(arr1,arr2)
````

## NumPy buffer  declaration

The `np.ndarray[np.float64_t, ndim=2]`(older version: `cnp.ndarray[double]`) declares a NumPy array _buffer_ object.  Cython knows how to interact with this array-like object efficiently.  The `np.float64_t` (older version: `double`) in square brackets is the (scalar) dtype of the array elements.

````python
%%cython

import numpy as np
cimport numpy as np

def mul_v2(np.ndarray[np.float64_t, ndim=2] x, np.ndarray[np.float64_t, ndim=2] y):

    cdef int m1 = x.shape[0]
    cdef int n1 = x.shape[1]
    cdef int m2 = y.shape[0]
    cdef int n2 = y.shape[1]
    cdef np.ndarray[np.float64_t, ndim=2] z = np.zeros([m1, n2], dtype=float)

    cdef int i
    cdef int j
    cdef int k

    cdef double value

    for i in range(m1):
        for j in range(n2):
            value = 0
            for k in range(m2):
                value += x[i][k] * y[k][j]
            z[i][j] = value

    return z
````

````python
%timeit -n 1 mul_v2(arr1,arr2)
````

### Typed memoryview 

- Typed memoryview allow efficient access to memory buffers, without any python overhead



![name](img/mem_view.png)

[Image source](https://github.com/kwmsmith/scipy-2015-cython-tutorial/blob/master/cython-scipy-2015-kurt-smith.pdf)

````python
%%cython
import numpy as np

narr = np.arange(27, dtype=np.dtype("i")).reshape((3, 3, 3))
cdef int [:, :, :] narr_view = narr


print("NumPy sum of the NumPy array before assignments: %s" % narr.sum())

# NumPy-style syntax for assigning a single value to all elements.
narr_view[:, :, :] = 3

print(f"NumPy sum of NumPy array after assignments: %s" % narr.sum())
````

````python
%%cython --annotate

def sum3d(int[:, :, :] arr):
    cdef size_t i, j, k, I, J, K
    cdef int total = 0
    I = arr.shape[0]
    J = arr.shape[1]
    K = arr.shape[2]
    for i in range(I):
        for j in range(J):
            for k in range(K):
                total += arr[i, j, k]
    return total
````

````python
%timeit -n 10000 narr.sum()
````

````python
%timeit -n 10000 sum3d(narr)
````

We can see that Cython version is faster than numpy version by comparing the above times. This is because `narr.sum()` involved:
- Python method call
- Some dispatch overhead

Hence, for this example, Cython is favored.

Declares `x` and `y` to be two dimensional contiguous typed memoryview 

````python
%%cython

import numpy as np
cimport numpy as np

DTYPE = np.float64
ctypedef np.float_t DTYPE_t

def mul_mv(double[:,:] x, double[:,:] y):

    cdef int m1 = x.shape[0]
    cdef int n1 = x.shape[1]
    cdef int m2 = y.shape[0]
    cdef int n2 = y.shape[1]

    cdef double[:,:] z = np.zeros([m1, n2], dtype=float)

    cdef int i
    cdef int j
    cdef int k

    cdef double value

    for i in range(m1):
        for j in range(n2):
            value = 0
            for k in range(m2):
                value += x[i,k] * y[k,j]
            z[i,j] = value

    return z
````

````python
%timeit -n 100 mul_mv(arr1,arr2)
````

### How fast is numpy?

````python
def mul_np(x,y):
    """
    Multiply two arrays using numpy.
    """
    return np.matmul(x,y)
````

````python
%timeit -n 100 mul_mv(arr1,arr2)
%timeit -n 100 mul_np(arr1,arr2)
````

Unlike the previous sum example, matrix multiplication favors numpy version because it calls **BLAS library**, which is highly optimized numerical libraries.

BLAS is:
- Hand-optimized C / assembly
- Uses SIMD (AVX)
- Uses cache blocking
- Uses multithreading
- Uses CPU-specific optimizations
