The problem is to simulate a random motion of $n$ objects over a discrete time. 

Concretely, there is:
- a unit square $[0,1]^2$, 
- $n$ points within the unit square, 
- and the time is discrete $t=0, 1, 2, \dots$. 

At time $t=0$, the positions of $n$ points are randomly and uniformly distributed within the unit square; call these positions $\{p_0, p_1, p_2,\dots, p_{n-1}\}$. 

At every time step $t \geq 0$, every point $i$, chooses to randomly move in one of four directions: left, right, up, down. The distance is also random uniform number on $[0, \delta]$, where $\delta$ is given. 

That is, at every time step $t$ and for every $i$ we generate a random move as: 
$$ p_i := p_i + r_i \cdot u_i$$
where 
$$ r_i \sim \text{Uniform}[0, \delta] $$
and 
$u_i$ represents a random direction, i.e. a randomly chosen vector among $(-1, 0), (1, 0), (0, -1), (0, 1)$.

**Dynamics**

Now, one would like to examine and plot the diagram of the minimum distance $d_{\min}$ among these $n$ points over $T$ iterations.

```python
import random

def main_original(n, delta, T):
    """
    n: is the number of uniformly at random generated points in the unit square
    delta: a maximal move of a point in one of four random directions: left, right, up, or down
    T: number of iterations
    return:
    lst_of_min_distances: of the minimum distances among all n points over times: t=0, 1, 2, \dots, T - 1,
    it is a list of reals of length T"""
    assert n >= 2

    dx = [0, 1, 0, -1]
    dy = [1, 0, -1, 0]
    points = [(random.uniform(0, 1), random.uniform(0, 1)) for _ in range(n)]
    min_dist = []

    def compute_min_dist(points):   # brute-force: compute distance of all pairs
        min_d = float('inf')
        for i in range(n):
            for j in range(i+1, n):
                x1, y1 = points[i]
                x2, y2 = points[j]
                d = ((x1-x2)**2 + (y1-y2)**2)**(1/2)
                min_d = min(min_d, d)

        return min_d

    min_dist.append(compute_min_dist(points))
    for _ in range(1, T):
        for i in range(n):
            x, y = points[i]
            dir_i = random.randint(0, 3)
            d = random.uniform(0, delta)

            x_n, y_n = x + d * dx[dir_i], y + d * dy[dir_i]
            points[i] = (x_n, y_n)

        min_dist.append(compute_min_dist(points))

    return min_dist
```

```python
import numpy as np

def main_optimized(n, delta, T):
    """
    n: is the number of uniformly at random generated points in the unit square
    delta: a maximal move of a point in one of four random directions: left, right, up, or down
    T: number of iterations
    return:
    lst_of_min_distances: of the minimum distances among all n points over times: t=0, 1, 2, \dots, T - 1,
    it is a list of reals of length T"""
    assert n >= 2

    dx = np.array([0, 1, 0, -1])
    dy = np.array([1, 0, -1, 0])
    points = np.random.rand(n, 2)
    min_dist = []

    # numpy version of min_dist (vectorized version)
    def compute_min_dist_np(points):
        diff = points[:, None, :] - points[None, :, :]   # shape (n, n, 2)
        # square distance of all pairwise points
        dist_sq = np.sum(diff**2, axis=2)                # shape (n, n)

        np.fill_diagonal(dist_sq, np.inf)                # ignore self-distance

        return np.sqrt(np.min(dist_sq))

    min_dist.append(compute_min_dist_np(points))
    for _ in range(1, T):
        dirs = np.random.randint(0, 4, size=n)
        dists = np.random.uniform(0, delta, size=n)

        points[:, 0] += dists * dx[dirs]
        points[:, 1] += dists * dy[dirs]

        min_dist.append(compute_min_dist_np(points))

    return min_dist
```

#### cf.) Changing min distance computing algorithm from brute-force to uniform grid approach

```python
import math

def main_optimized_grid(n, delta, T):
    """
    n: is the number of uniformly at random generated points in the unit square
    delta: a maximal move of a point in one of four random directions: left, right, up, or down
    T: number of iterations
    return:
    lst_of_min_distances: of the minimum distances among all n points over times: t=0, 1, 2, \dots, T - 1,
    it is a list of reals of length T"""
    assert n >= 2

    dx = [0, 1, 0, -1]
    dy = [1, 0, -1, 0]
    points = [(random.uniform(0, 1), random.uniform(0, 1)) for _ in range(n)]
    min_dist = []

    # uniform grid method to find min distance among all pairs of points
    def compute_min_dist_grid(points):
        # --- Step 0: find initial best with brute-force using a small subset ---
        m = min(n, 50)
        min_d = float("inf")
        for i in range(m):
            for j in range(i+1, m):
                x1, y1 = points[i]
                x2, y2 = points[j]
                d = math.hypot(x1 - x2, y1 - y2)
                min_d = min(min_d, d)

        if min_d == 0.0: return 0.0

        # keep shrinking cell size if we find a smaller min_d
        cell_sz = min_d
        while True:
            grid = {}  # (cx,cy) -> list of indices
            improved = False
            min_d_thres = min_d

            for i, (x, y) in enumerate(points):
                cx = math.floor(x / cell_sz)
                cy = math.floor(y / cell_sz)

                # Step 2: check 9 neighbor cells
                for dx in (-1, 0, 1):
                    for dy in (-1, 0, 1):
                        cell = (cx + dx, cy + dy)
                        if cell in grid:
                            for j in grid[cell]:
                                x2, y2 = points[j]
                                d = math.hypot(x - x2, y - y2)
                                min_d_thres = min(min_d_thres, d)

                # Step 3: insert point
                grid.setdefault((cx, cy), []).append(i)

            # Step 4: shrink cell size if improved
            if min_d_thres < min_d:
                min_d = min_d_thres
                if min_d == 0.0: return 0.0

                cell_sz = min_d
                improved = True

            if not improved:
                return min_d


    min_dist.append(compute_min_dist_grid(points))
    for _ in range(1, T):
        for i in range(n):
            x, y = points[i]
            dir_i = random.randint(0, 3)
            d = random.uniform(0, delta)

            x_n, y_n = x + d * dx[dir_i], y + d * dy[dir_i]
            points[i] = (x_n, y_n)

        min_dist.append(compute_min_dist_grid(points))

    return min_dist
```
