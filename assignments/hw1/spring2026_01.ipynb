# Problem 1. (30pt)

Write a function f(n) to calculate the number of ways of representing $n$ as a sum of 1, 2, and 5, where the order of summands is important. For example: 

---

n = 1  
1 = 1  
f(1) = 1  

---

n = 2  
2 = 1 + 1  
2 = 2  
f(2) = 2  

---

n = 3  
3 = 1 + 1 + 1  
3 = 1 + 2  
3 = 2 + 1  
f(3) = 3 

---

n = 5  
5 = 1 + 1 + 1 + 1 + 1  
5 = 1 + 1 + 1 + 2  
5 = 1 + 1 + 2 + 1  
5 = 1 + 2 + 1 + 1  
5 = 2 + 1 + 1 + 1  
5 = 2 + 2 + 1  
5 = 2 + 1 + 2  
5 = 1 + 2 + 2  
5 = 5  
f(5) = 9

---

(10pt) Write a recursive solution f_rec() without memoization and time it for $n=10, 25$.  

(10pt) Write a recursive solution with memoization or a wrapper function f_memo() and time it for $n=10, 25, 50, 100$.  

(10pt) Write an iterative solution f_it() and time it for $n=10, 25, 50, 100$.  



```python
def f_rec(n):
    if n < 0:
        return 0
    elif n == 0:
        return 1
    else:
        return f_rec(n-1) + f_rec(n-2) + f_rec(n-5)
```

```python
%timeit -r 10 -n 20 f_rec(10)
%timeit -r 10 -n 20 f_rec(25)
```

```python
from functools import lru_cache

@lru_cache(maxsize=128)
def f_memo(n):
    if n < 0:
        return 0
    elif n == 0:
        return 1
    else:
        return f_memo(n-1) + f_memo(n-2) + f_memo(n-5)
```

```python
%timeit -r 10 -n 20 f_memo(10)
%timeit -r 10 -n 20 f_memo(25)
%timeit -r 10 -n 20 f_memo(50)
%timeit -r 10 -n 20 f_memo(100)
```

```python
def f_it(n):
    dp = [0] * (n + 2)
    dp[0] = 0
    dp[1] = 1

    for i in range(2, n + 2):
        dp[i] = dp[i-1] + dp[max(0, i-2)] + dp[max(0, i-5)]

    return dp[n+1]
```

```python
%timeit -r 10 -n 20 f_it(10)
%timeit -r 10 -n 20 f_it(25)
%timeit -r 10 -n 20 f_it(50)
%timeit -r 10 -n 20 f_it(100)
```
